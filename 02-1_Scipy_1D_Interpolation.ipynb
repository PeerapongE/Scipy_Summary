```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
from scipy.interpolate import interp1d

%matplotlib inline
```

```python
x_cont = np.linspace(0,1,1000)
y_cont = np.sin(2 * np.pi * x_cont)
```

```python
x = np.linspace(0, 1, 8)
x
```

```python
y = np.sin(2 * np.pi * x)
y
```

```python
# Create relationship (linear as default)
linear_interp = interp1d(x, y)
```

```python
x_est = np.linspace(0, 1, 50)
x_est
```

```python
y_est_linear = linear_interp(x_est)
y_est_linear
```

```python
plt.plot(x_cont,y_cont,'-k')
plt.plot(x_est,y_est_linear,'-')
```

## Various Method for Interploation using Scipy

```python
# Create relationship (cubic interpolation)


nearest_interp   = interp1d(x,y, kind='nearest')
previous_interp  = interp1d(x,y, kind='previous')
next_interp      = interp1d(x,y, kind='next')


cubic_interp     = interp1d(x,y, kind='cubic')
quadratic_interp = interp1d(x,y, kind='quadratic')
```

```python

y_est_nearest   = nearest_interp (x_est)
y_est_previous  = previous_interp(x_est)
y_est_next      = next_interp(x_est)

y_est_cubic     = cubic_interp(x_est)
y_est_quadratic = quadratic_interp (x_est)
```

```python
# plt.plot(x_cont,y_cont,'-')
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.interpolate.interp1d.html#scipy.interpolate.interp1d
plt.plot(x, y,'ok')
plt.plot(x_cont,y_cont,'-k')

plt.plot(x_est,y_est_nearest,'-')
plt.plot(x_est,y_est_previous,'-')
plt.plot(x_est,y_est_next,'-')

plt.title('Interpolation Method by Neighbour')
plt.legend(['True_Point','True_Line','Nearest','Previous','Next',])
```

```python

plt.plot(x, y,'ok')
plt.plot(x_cont,y_cont,'-k')


plt.plot(x_est,y_est_cubic,'-')
plt.plot(x_est,y_est_quadratic,'-')
plt.plot(x_est,y_est_linear,'-')

plt.title('Interpolation Method by Function Estimation')
plt.legend(['True_Point','True_Line','Cubic','Quadratic','Linear'])

```
